## Rank/Degree of tensor
Another word for rank is degree, these term in tensorflow means the number of dimensions in each tensors. The rank or degree of what we created above is 0.Since there is no dimensions.                                              
Now we'll create some tensors of higher dimensions below:-

```python
rank1_tensor = tf.Variable([12,34,43,], tf.string)
rank2_tensor = tf.Variable([["Tushar","Anupama","Tanu"],
                            ["Raj","Deep","Raju"]], tf.string) 
rank3_tensor = tf.Variable([[[12],[24]],
                            [[34],[43]],
                            [[29],[98]],
                            [[23],[54]],
                            [[65],[35]]], tf.int16)
```

To determine the rank of any tensor we call the following method 

```python
tf.rank(rank1_tensor)
```

```python
tf.rank(rank3_tensor)
```

## Changing shapes 
The number of elements present in a Tensor is the product of the sizes of all shapes. There are often many shapes with same number of elements making it easy to be able to change the shape of it.                                                                                                          
The example below shows how to change the shape of tensors

```python
tensor1 = tf.ones([1,2,3])
print(tensor1)
```

```python
tensor2 = tf.reshape(tensor1, [1,1,6])
```

```python
print(tensor2)
```

```python
tensor3 = tf.reshape(tensor1, [1,2,3])
```

```python
print(tensor3)
```

```python
tensor4 = tf.reshape(tensor1, [3, -1]) #-1 tells the tensor to fill the gap by calculating the size of dimension in that place.    
                                       #like her it's(3,2).
```

```python
print(tensor4)
```
